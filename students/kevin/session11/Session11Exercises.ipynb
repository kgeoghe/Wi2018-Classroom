Find most danceable tracks, sorted by danceability

```python
music.loc[(music.danceability > 0.8) & (music.loudness < -5)].sort_values(['danceability'], ascending=False)
```

Find top 5 most danceable tracks

```python
music.loc[(music.danceability > 0.8) & (music.loudness < -5)].sort_values(['danceability'], ascending=False).head(5)
```
